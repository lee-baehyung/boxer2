## [문제 24] 문자열에서 특정 단어 찾기
### `"Lumos"`와 `"Maxima"`를 교차해 "LMuaoxsima"로 조합하세요.

```python
spell = "Lumos Maxima"
words = spell.split()
word1 = words[0]
word2 = words[1]
spell2 = ""
for i in range(min(len(word1), len(word2))):
    spell2 += word1[i] + word2[i]

if len(word1) > len(word2):
    spell2 += word1[len(word2):]
elif len(word2) > len(word1):
    spell2 += word2[len(word1):]

print(spell2)
```

## [문제 25] 문자열 압축
### `"aaabbbccc"`를 "a3b3c3" 형식으로 압축하세요.

```python
def compress_string(s):
    result = ""
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            result += s[i - 1] + str(count)
            count = 1
    result += s[-1] + str(count)
    return result

compressed = compress_string("aaabbbccc")
print(compressed)
```

## [문제 27] 중복 문자 제거
### `"magic spells"`에서 중복된 문자를 제거하여 출력하세요.

```python
def remove_duplicates(s):
    result = ""
    for char in s:
        if char not in result:
            result += char
    return result

s = "magic spells"
print(remove_duplicates(s))

# 답이 magic pe 나와야 되는게 아닌가..?
```
