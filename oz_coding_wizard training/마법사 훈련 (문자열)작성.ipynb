## [ë¬¸ì œ 24] ë¬¸ìì—´ì—ì„œ íŠ¹ì • ë‹¨ì–´ ì°¾ê¸°
### `"Lumos"`ì™€ `"Maxima"`ë¥¼ êµì°¨í•´ "LMuaoxsima"ë¡œ ì¡°í•©í•˜ì„¸ìš”.

```python
spell = "Lumos Maxima"
words = spell.split()
word1 = words[0]
word2 = words[1]
spell2 = ""
for i in range(min(len(word1), len(word2))):
    spell2 += word1[i] + word2[i]

if len(word1) > len(word2):
    spell2 += word1[len(word2):]
elif len(word2) > len(word1):
    spell2 += word2[len(word1):]

print(spell2)
```

## [ë¬¸ì œ 25] ë¬¸ìì—´ ì••ì¶•
### `"aaabbbccc"`ë¥¼ "a3b3c3" í˜•ì‹ìœ¼ë¡œ ì••ì¶•í•˜ì„¸ìš”.

```python
def compress_string(s):
    result = ""
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            result += s[i - 1] + str(count)
            count = 1
    result += s[-1] + str(count)
    return result

compressed = compress_string("aaabbbccc")
print(compressed)
```
